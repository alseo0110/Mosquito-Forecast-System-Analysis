csv 파일 읽기

```python
import pandas as pd

df = pd.read_csv('/content/기온.csv', encoding='EUC-KR')

# df.head()

df.info()
```

결측치 찾기

```python
df[df.isnull().any(axis=1)]
```

```python
df.isna().sum()
```

결측치 채우고 필요없는 컬럼 삭제하기

```python
df.loc[df['일시'] == '2022-08-08', '최저기온(℃)'] = 25.2
df.loc[df['일시'] == '2022-08-08', '일교차'] = 3.2

df = df.dropna()

df = df.drop(columns=['\t\t지점번호', '지점명', '\t최고기온시각', '최저기온시각'])
```

```python
df.head()
```

```python
# 일시 컬럼 datetime 변환
df['일시'] = pd.to_datetime(df['일시'], errors='coerce')

# 연도 필터 (2021~2025)
df = df[df['일시'].dt.year.between(2021, 2025)]

# 월 필터 (5월~10월)
df = df[df['일시'].dt.month.isin([5,6,7,8,9,10])]
```

저장

```python
df.to_csv('/content/기온_수정.csv', index = False, encoding="EUC-KR")
```
